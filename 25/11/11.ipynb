```python
with open("input.txt", "r") as file:
    content = file.read().splitlines()
```

```python
cons = dict()
for line in content:
    a, b = line.split(": ")
    b = b.split(" ")
    cons[a] = b
```

```python
# top down dp
# go from back to front and save per node how many nodes to "out"
dp = dict()
dp["out"] = 1


def dfs(node):
    if node in dp:
        return dp[node]
    total = 0
    for nc in cons[node]:
        total += dfs(nc)
    dp[node] = total
    return total


dfs("you")
```

### Part 2

```python
dp = dict()
dp["out"] = {(False, False): 1}  # save if passed dac and fft: save count


def dfs2(node):
    if node in dp:
        return dp[node]
    total = dict()
    for nc in cons[node]:
        res = dfs2(nc)
        for key, val in res.items():
            if node == "dac":
                key = (True, key[1])
            if node == "fft":
                key = (key[0], True)
            if key in total:
                total[key] += val
            else:
                total[key] = val
    dp[node] = total
    return total


dfs2("svr")[(True, True)]
```
