```python
with open("input.txt", "r") as file:
    ranges, ids = file.read().split("\n\n")
    ranges = [
        range(s, e + 1)
        for s, e in (map(int, r.split("-")) for r in ranges.splitlines())
    ]
    ids = list(map(int, ids.splitlines()))
```

### Part 1

```python
fresh = 0
for id in ids:
    for r in ranges:
        if id in r:
            fresh += 1
            break
fresh
```

### Part 2

```python
ranges.sort(key=lambda r: r.start)
unique_ids = 0
last_right = -1
for r in ranges:
    if r.start > last_right:
        unique_ids += r.stop - r.start
        last_right = r.stop
    elif r.stop > last_right:
        unique_ids += r.stop - last_right
        last_right = r.stop
unique_ids
```
