### Input

```python
with open("input.txt", "r") as f:
    lines = f.readlines()
```

```python
def parse_line(line):
    line = line.strip()
    return line[0], int(line[1:])
```

### Task 1

```python
state = 50
ans = 0
for l in lines:
    d, n = parse_line(l)
    vz = -1 if d == "L" else +1
    state = (state + vz * n) % 100
    if state == 0:
        ans += 1
ans
```

### Task 2

```python
ans = 0
state = 50
for l in lines:
    d, n = parse_line(l)
    pstate = state
    vz = -1 if d == "L" else +1
    div, state = divmod(state + vz * n, 100)
    ans += abs(div)
    if d == "L":
        if pstate == 0:
            ans -= 1
        if state == 0:
            ans += 1
ans
```
