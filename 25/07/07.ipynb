```python
import re
from collections import defaultdict

with open("input.txt") as file:
    lines = file.read().splitlines()
start = lines[0].find("S")
```

```python
beams = defaultdict(int)
beams[start] = 1
total = 0
for l in lines:
    matches = list(re.finditer(r"\^", l))
    if len(matches) == 0:
        continue
    new_beams = beams.copy()
    for match in matches:
        if match.span()[0] in beams:
            count = beams[match.span()[0]]
            if count == 0:
                continue
            total += 1
            new_beams[match.span()[0] - 1] += count
            new_beams[match.span()[0]] -= count
            new_beams[match.span()[0] + 1] += count
    beams = new_beams
sum(beams.values()), total
```
