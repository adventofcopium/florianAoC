```python
import numpy as np

opmap = {
    "+": np.add,
    "*": np.multiply,
}

with open("input.txt") as f:
    lines = f.readlines()

lines_clean = [l.strip().split() for l in lines]
```

### Part 1

```python
ops = lines_clean[-1]
numbers = np.array(lines_clean[:-1]).astype(int).T

result = 0

for i, op in enumerate(ops):
    func = opmap[op]
    result += func.reduce(numbers[i])

print(result)
```

### Part 2

```python
import re

# do rle of ops ((*,1),( ,4),(+,1),( ,7),...)
ops = lines[-1].strip("\n") + " "
ops = re.findall(r"(\+|\*| +)", ops)
ops = list(zip(ops[::2], map(len, ops[1::2])))

# accessable string matrix [row, col]
numbers = np.array([list(c.strip("\n")) for c in lines[:-1]])

num_rows = len(lines_clean)

col = 0
result = 0
for op, ln in ops:
    func = opmap[op]
    nums = numbers[:, col : col + ln].T
    nums = list(map("".join, nums))
    nums = list(map(int, nums))
    result += func.reduce(nums)
    col += ln + 1

print(result)
```
