```python
with open("input.txt", "r") as f:
    s = f.read()

ranges = s.split(",")
```

```python
def test_range_inner(lower, upper, k=2):
    ans = set()
    cur = str(lower)
    ln = len(cur)
    while ln % k != 0:
        cur = "1" + "0" * ln
        ln = len(cur)
    # clamp lower
    cur = str(max(lower, int(cur)))

    first = cur[: ln // k]
    while True:
        cur = int(first * k)
        if cur > upper:
            break
        if cur < lower:
            first = str(int(first) + 1)
            continue
        ans.add(cur)
        first = str(int(first) + 1)
    return ans


def test_range(rng, k=2):
    lower, upper = rng.split("-")
    ans = set()
    upper = int(upper)
    lower = int(lower)

    if k == None:  # do all ks
        for k in range(2, len(str(upper)) + 1):
            s = test_range_inner(lower, upper, k)
            ans |= s
    else:
        ans |= test_range_inner(lower, upper, k)

    return sum(ans)
```

```python
ans = 0
for r in ranges:
    ans += test_range(r, k=2)
ans
```
