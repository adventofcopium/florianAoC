```python
import numpy as np

with open("input.txt") as f:
    lines = f.readlines()

points = [line.strip().split(",") for line in lines]
points = np.array(points, dtype=int)
```

```python
diffs = points[:, np.newaxis, :] - points[np.newaxis, :, :]
dists = np.sqrt(np.sum(diffs**2, axis=-1))
i_idx, j_idx = np.triu_indices(len(points), k=1)  # get indices of all relevant entries in the upper triangle
upper_dists = dists[i_idx, j_idx]
```

```python
def find(x, parent):
    if parent[x] == x:
        return x
    parent[x] = find(parent[x])
    return parent[x]


def union(x, y, parent):
    rootX = find(x, parent)
    rootY = find(y, parent)
    if rootX != rootY:
        parent[rootY] = rootX
        return True
    return False
```

### Part 1

```python
k = 1000
if k > len(upper_dists):
    k = len(upper_dists)
smallest_indices = np.argpartition(upper_dists, k)[:k]

top_k = []
for index in smallest_indices:
    p1 = i_idx[index]
    p2 = j_idx[index]
    top_k.append((int(p1), int(p2)))

parent = [i for i in range(len(points))]

for p1, p2 in top_k:
    union(p1, p2, parent)

clusters = {}
for i in range(len(points)):
    root = find(i, parent)
    if root not in clusters:
        clusters[root] = []
    clusters[root].append(i)

lengs = sorted([len(c) for c in clusters.values()])[-3:]
ans = int(np.prod(lengs))
ans
```

### Part 2

```python
sorted_idx = upper_dists.argsort()
j_s = j_idx[sorted_idx]
i_s = i_idx[sorted_idx]
parent = [i for i in range(len(points))]
num_clusters = len(points)
for idx in range(len(i_s)):
    p1 = i_s[idx]
    p2 = j_s[idx]
    if union(p1, p2, parent):
        num_clusters -= 1
    if num_clusters == 1:
        break
last_point1 = points[i_s[idx]]
last_point2 = points[j_s[idx]]
int(last_point1[0] * last_point2[0])
```
