```python
def get_matrix(file):
    m = Matrix.from_file(file)
    m = np.pad(m.data, pad_width=1, mode="constant", constant_values=".")
    return m
```

```python
def add_rings(m, pos, factor=1):
    for r, c in zip(pos[0], pos[1]):
        for dr in [-1, 0, 1]:
            for dc in [-1, 0, 1]:
                if dr == 0 and dc == 0:
                    continue
                m[r + dr, c + dc] += factor
    return m
```

```python
def get_count(m):
    count = np.zeros_like(m, dtype=int)
    rolls = np.where(m == "@")
    add_rings(count, rolls, 1)
    return count
```

```python
def get_valid(m, count):
    return np.where((m == "@") & (count < 4))
```

```python
ans = 0
m = get_matrix("input.txt")
count = get_count(m)
while True:
    valid = get_valid(m, count)
    valid_cnt = len(valid[0])
    if valid_cnt == 0:
        break
    ans += valid_cnt
    count = add_rings(count, valid, -1)
    m[valid[0], valid[1]] = "."
ans
```
