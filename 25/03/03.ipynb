```python
import numpy as np

with open("input.txt", "r") as file:
    lines = [[int(x) for x in line] for line in file.read().splitlines()]
```

```python
def find_max_indices(lines, k=2):
    # find max [0:-k] and set as max_k
    # find max [max_k:-k+1] and set as max_k+1
    # ...
    # num = [10**k * max_k for k in range(k)]
    ans = 0

    for line in lines:
        max_ks = [-1]  # save indices
        for k_ in range(k - 1, -1, -1):
            left = max_ks[-1] + 1
            right = -k_ if k_ != 0 else None
            max_k = np.argmax(line[left:right])
            max_ks.append(max_k + left)
        max_ks = max_ks[1:][::-1]

        ans += sum([10**k * line[max_ks[k]] for k in range(len(max_ks))])

    return ans
```
